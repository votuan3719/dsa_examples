```python
class Node():
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
```

### 102. Binary Tree Level Order Traversal

#### Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

```python
root = Node(3)
nine = Node(9)
twenty = Node(20)
fifteen = Node(15)
seven = Node(7)

# left subtree
root.left = nine

# right subtree
root.right = twenty
twenty.left = fifteen
twenty.right = seven
```

```python
from collections import deque

def levelOrder(root):
    levels = []
    
    queue = deque([root])
    while queue:
        level = []
        
        for _ in range(len(queue)):
            node = queue.popleft()
            level.append(node.value)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
        levels.append(level)
    return levels
    
```

```python
levelOrder(root)
```

### 1161. Maximum Level Sum of a Binary Tree

#### Given the root of a binary tree, the level of its root is 1, the level of its children is 2, and so on.

#### Return the smallest level x such that the sum of all the values of nodes at level x is maximal.

```python
root = Node(1)
seven = Node(7)
zero = Node(0)
_seven = Node(7)

# left subtree
root.left = seven
seven.left = _seven
seven.right = Node(-8)

# right subtree
root.right = zero
```

```python
from collections import deque

def maxLevelSum(root):
    max_lvl, max_sum = 0, 0
    
    if root is None:
        return 0
    
    curr_lvl = 0
    queue = deque([root])
    while queue:
        curr_lvl += 1     
        lvl_sum = 0
        for _ in range(len(queue)):
            node = queue.popleft()
            lvl_sum += node.value

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
        if lvl_sum > max_sum:
            max_lvl, max_sum = curr_lvl, lvl_sum      
    return max_lvl
```

```python
maxLevelSum(root)
```

### 637. Average of Levels in Binary Tree

#### Given the root of a binary tree, return the average value of the nodes on each level in the form of an array. 
 

```python
root = Node(3)
nine = Node(9)
twenty = Node(20)
fifteen = Node(15)
seven = Node(7)

# left subtree
root.left = nine

# right subtree
root.right = twenty
twenty.left = fifteen
twenty.right = seven
```

```python
from collections import deque

def averageOfLevels(root):
    levels = []
    
    queue = deque([root])
    while queue:
        sum_lvl = 0
        num_nodes = len(queue)
        
        for _ in range(num_nodes):
            node = queue.popleft()
            sum_lvl += node.value
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
            
        average = sum_lvl/num_nodes
        levels.append(average)
        
    return levels
    
```

```python
averageOfLevels(root)
```
