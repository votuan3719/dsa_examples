```python
def minDepth(node):
    # if a node has no children, return 0
    if node == None:
        return 0
    
    # check if a node is a leaf 
    if node.left == None:
        depth = minDepth(node.right)
    elif node.right == None:
        depth = minDepth(node.left)
    else:
        # recursively call the current node's children
        # every node has its own unique left and right values
        left = minDepth(node.left)
        right = minDepth(node.right)
        depth = min(left, right)
    
    # add one to the depth
    return 1 + depth

minDepth(root)
```

```python
def minDepth(node):
    if not node:
        return 0

    left = minDepth(node.left)
    right = minDepth(node.right)
    
    return 1 + min(left, right)

minDepth(root)
```

```python
def minDepth(root):
    def dfs(node, min_depth):
        if node.left == None and node.right == None:
            return min_depth
        
        left = dfs(node.left, min_depth + 1)
        right = dfs(node.right, min_depth + 1)
        return min(left, right)

    return dfs(root, 1)

minDepth(root)
```

```python
def goodNodes(root):
    def dfs(node, max_so_far):
        num_nodes = 0
        if not node:
            return 0
        
        if node.value >= max_so_far:
            num_nodes += 1
        
        left = dfs(node.left, max(max_so_far, node.value))
        right = dfs(node.right, max(max_so_far, node.value))
        num_nodes += left + right
        
        return num_nodes 
        
    return dfs(root, float('-inf'))

goodNodes(root)
```

```python
def goodNodes(root):
    num_nodes = 0
    if not root:
        return 0
    
    stack = [(root, float('-inf'))]
    while stack:
        node, max_so_far = stack.pop()
        
        if node.value > max_so_far:
            num_nodes += 1
            max_so_far = node.value
        
        if node.left:
            stack.append((node.left, max_so_far))
        if node.right:
            stack.append((node.right, max_so_far))
            
    return num_nodes

goodNodes(root)
```

```python
def lowestCommonAncestor(root, p, q):
    def dfs(node):
        if not node:
            return None
        
        if node == p or node == q:
            return node
        
        left = dfs(node.left)
        right = dfs(node.right)
        
        if left and right:
            return node
        return left or right
    return dfs(root)
    
ans = lowestCommonAncestor(root, six, two)
```

```python
def diameterOfABinaryTree(root):
    max_diameter = 0
    def dfs(node):
        nonlocal max_diameter
        if not node:
            return 0
        
        left = dfs(node.left)
        right = dfs(node.right)
        
        max_diameter = max(max_diameter, left + right) 
        return max(left, right) + 1
                     
    dfs(root)
    return max_diameter

diameterOfABinaryTree(root)
```

```python
from collections import deque

def zigzagLevelOrder(root):
    ans = []
    queue = deque([root])
    
    traverse_left = False
    
    if not root:
        return
    
    while queue:
        level = []
        for _ in range(len(queue)):
            node = queue.popleft()
            level.append(node.value)
            
            if not traverse_left:
                if node.right:
                    queue.append(node.right)
                if node.left:
                    queue.append(node.left)
                traverse_left = True
            else:
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                traverse_left = False
        ans.append(level)
    
    return ans

zigzagLevelOrder(root)
    
```

```python
from collections import deque

def rangeSumBST(root, low, high):
    queue = deque([root])
    
    curr_sum = 0
    while queue:
        for _ in range(len(queue)):
            node = queue.popleft()
            if node.value >= low and node.value <= high:
                curr_sum += node.value
            
            if node.left and node.value > low:
                queue.append(node.left)
            if node.right and node.value < high:
                queue.append(node.right)
    
    return curr_sum

rangeSumBST(root, 7, 15)
```

```python
from collections import deque

def getMinimumDifference(root):
    min_diff = float('inf')
    
    stack = []
    prev, curr = None, root
    while stack or curr:
        while curr:
            curr = curr.left
        
        prev = curr
        curr = curr.right
    return min_diff
```
