#### “Make sequence smallest/largest” (removing k digits, most competitive, etc.)

```python
def monotonic_stack(arr, increasing=True):
    # if increasing=True, stack keeps values in increasing order
    stack = []
    for x in arr:
        # while top violates monotonicity with x, pop it
        while stack and ((x < stack[-1]) if increasing else (x > stack[-1])):
            stack.pop()
        stack.append(x)
    return stack
```

```python
from collections import deque

def most_competitive(nums, k):
    stack = deque()
    to_remove = len(nums) - k
    for x in nums:
        # while we can remove and top > x → pop for a smaller sequence
        while stack and to_remove and stack[-1] > x:
            stack.pop()
            to_remove -= 1
        stack.append(x)
    # if still need to remove extras, drop from end
    while to_remove:
        stack.pop()
        to_remove -= 1
    # only keep first k
    return list(stack)[:k]

# Example
print(most_competitive([3,5,2,6], 2))  # → [2,6]
```

```python
def next_greater(nums):
    res = [-1]*len(nums)
    stack = []  # will store indices whose next greater we haven’t found
    for i, x in enumerate(nums):
        # whenever current x > nums[stack[-1]], we found next greater for stack[-1]
        while stack and x > nums[stack[-1]]:
            j = stack.pop()
            res[j] = x
        stack.append(i)
    return res

# Example
print(next_greater([2, 1, 2, 4, 3]))  # → [4, 2, 4, -1, -1]
```
