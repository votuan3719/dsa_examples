## Checking for existence

```python
def containsDuplicate(nums):
    seen = set()
    
    for val in nums:
        if val in seen:
            return True
        seen.add(val)
    
    return False
```

```python
nums = [1,2,3,1]
```

```python
containsDuplicate(nums)
```

```python
def destCity(paths):
    seen = set()
    for path in paths:
        departure, _ = path
        seen.add(departure)
    
    for path in paths:
        _, destination = path
        if destination not in seen:
            return destination
    
    return ""
        
```

```python
paths = [["London","New York"],["New York","Lima"],["Lima","Sao Paulo"]]
```

```python
destCity(paths)
```

```python
def isPathCrossing(path):
    seen = set()
    direction = {
        'N': (0, 1),
        'S': (0, -1),
        'E': (1, 0),
        'W': (-1, 0)
    }
    
    x, y = 0, 0
    seen.add((x,y))
    
    for c in path:
        dx, dy = direction[c]
        x += dx
        y += dy
        
        if (x,y) in seen:
            return True
        seen.add((x,y))
    
    return False
```

```python
isPathCrossing('NESWW')
```

## Counting

```python
def sumOfUniques(nums):
    seen = set()
    
    for val in nums:
        if val in seen:
            seen.remove(val)
        else:
            seen.add(val)
    
    return sum(seen)
```

```python
nums = [1,2,3,4,5]
```

```python
sumOfUniques(nums)
```

```python
def maxFrequencyElements(nums):
    freq = {}
    count = 0
    
    for val in nums:
        if val in freq:
            freq[val] += 1
        else:
            freq[val] = 1
            
    max_freq = max(freq.values())
    for val in freq:
        if freq[val] == max_freq:
            count += 1
    
    return count * max_freq
```

```python
def maxFrequencyElements(nums):
    freq = {}
    
    for val in nums:
        if val in freq:
            freq[val] += 1
        else:
            freq[val] = 1
    
    buckets = [[] for _ in range(len(nums) + 1)]
    for val, freq in freq.items():
        buckets[freq].append(val)
    
    for f in range(len(nums), 0, -1):
        if buckets[f]:
            return len(buckets[f]) * f
    
    return -1
```

```python
nums = [1,2,2,3,1,4]
```

```python
maxFrequencyElements(nums)
```

```python
def findLucky(arr):
    freq = {}
    lucky = float('-inf')
    
    for val in arr:
        if val in freq:
            freq[val] += 1
        else:
            freq[val] = 1
    
    for val in freq:
        if freq[val] == val:
            lucky = max(lucky, val)
    
    return lucky if lucky != float('-inf') else -1
```

```python
arr = [2,2,2,3,3]
```

```python
findLucky(arr)
```

```python
def uniqueOccurrences(arr):
    freq = {}
    for val in arr:
        if val in freq:
            freq[val] += 1
        else:
            freq[val] = 1
    
    seen = set()
    for val in freq:
        if freq[val] in seen:
            return False
        seen.add(freq[val])
    
    return True
    
```

```python
arr = [1,2,2,1,1,3]
```

```python
uniqueOccurrences(arr)
```

```python
def frequencySort(s):
    freq = {}
    ans = []
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    
    seen = set(s)
    max_freq = max(freq.values())
    while max_freq > 0:
        for char in seen:
            if freq[char] == max_freq:
                ans.append(char * max_freq)
        max_freq -= 1
    
    return ''.join(ans)
```

```python
def frequencySort(s):
    freq = {}
    ans = []
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    
    buckets = [[] for _ in range(len(s) + 1)]
    for char, freq in freq.items():
        buckets[freq].append(char)
    
    for freq in range(len(s), 0, -1):
        for char in buckets[freq]:
            ans.append(char * freq)
    
    return ''.join(ans)
```

```python
def maxSubarrayLength(nums, k):
    freq = {}
    
    left = 0
    max_len = 0
    for right in range(len(nums)):
        if nums[right] in freq:
            freq[nums[right]] += 1
        else:
            freq[nums[right]] = 1
            
        while freq[nums[right]] > k:
            freq[nums[left]] -= 1
            left += 1
        max_len = max(max_len, right - left + 1)

    return max_len
    
```

```python
nums = [5,5,5,5,5,5,5]
k = 4
```

```python
maxSubarrayLength(nums, k)
```

```python
from collections import defaultdict

def numIdenticalPairs(nums):
    freq = defaultdict(int)
    ans = 0
    for val in nums:
        ans += freq[val]
        freq[val] += 1
    
    return ans
        
    
```

```python
nums = [1,2,3,1,1,3]
```

```python
numIdenticalPairs(nums)
```
