## Stack

```python
def removeStars(s):
    ans = []
    
    for char in s:
        if char == '*':
            ans.pop()
        else:
            ans.append(char)
    
    return ''.join(ans)

s = "erase*****"
removeStars(s)
```

```python
class MyQueue:
    def __init__(self):
        self.s1 = []
        self.s2 = []
    
    def push(self, x):
        self.s1.append(x)
    
    def pop(self):
        if len(self.s2) == 0:
            self.peek()
        return self.s2.pop()
    
    def peek(self):
        while len(self.s1) > 0:
            self.s2.append(self.s1.pop())
        return self.s2[-1]
    
    def empty(self):
        if len(self.s1) == 0 and len(self.s2) == 0:
            return True
        return False

queue = MyQueue()
queue.push(1)
queue.push(2)
queue.push(3)    
```
