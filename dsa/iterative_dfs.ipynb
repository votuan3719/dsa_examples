```python
class Node():
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
```

### Preorder (root -> left -> right)

#### Process immediately when popped.

#### Process parent before children.

```python
def preorder_iterative(root):
    if not root:
        return []
    
    result = []
    stack = [root]
    
    while stack:
        node = stack.pop()
        result.append(node.val)
        
        # Push right first, then left (stack is LIFO)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    
    return result
```

### Inorder (left -> root -> right)

#### Go left first, process when no left child.

#### Process in sored order (for BST)

```python
def inorder_iterative(root):
    result = []
    stack = []
    current = root
    
    while stack or current:
        # Go to leftmost node
        while current:
            stack.append(current)
            current = current.left
        
        # Process node
        current = stack.pop()
        result.append(current.val)
        
        # Move to right subtree
        current = current.right
    
    return result
```

```python
def successor(node):
    node = node.right
    while node.left:
        node = node.left
    return node

def predecessor(node):
    node = node.left
    while node.right:
        node = node.right
    return node

def deleteNode(node, target):
    if not node:
        return None
    
    if node.value > target:
        node.left = deleteNode(node.left, target)
    elif node.value < target:
        node.right = deleteNode(node.right, target)
    else:
        if not (node.left or node.right):
            node = None
        elif not node.right:
            pred = predecessor(node)
            node.value = pred.value
            node.left = deleteNode(node.left, pred.value)
        else:
            succ = successor(node)
            node.value = succ.value
            node.right = deleteNode(node.right, succ.value)

    return node
```

### Inorder Traversal Uses

```python
# 1. BST Validation
def isValidBST(root):
    def inorder(node):
        if not node:
            return []
        return inorder(node.left) + [node.val] + inorder(node.right)
    
    values = inorder(root)
    return values == sorted(values) and len(values) == len(set(values))

# 2. Kth Smallest Element in BST
def kthSmallest(root, k):
    stack = []
    current = root
    count = 0
    
    while stack or current:
        while current:
            stack.append(current)
            current = current.left
        
        current = stack.pop()
        count += 1
        if count == k:
            return current.val
        
        current = current.right

# 3. BST to Sorted Array/List
def bstToSortedArray(root):
    result = []
    def inorder(node):
        if node:
            inorder(node.left)
            result.append(node.val)
            inorder(node.right)
    
    inorder(root)
    return result
```

### Postorder (left -> right -> root)

#### Visit twice.

#### Process children before parent.

```python
def postorder_one_stack(root):
    if not root:
        return []
    
    result = []
    stack = [(root, False)]
    
    while stack:
        node, visited = stack.pop()
        
        if visited:
            # Second visit: process node
            result.append(node.val)
        else:
            # First visit: push in reverse order
            stack.append((node, True))
            if node.right:
                stack.append((node.right, False))
            if node.left:
                stack.append((node.left, False))
    
    return result
```

### Postorder Traversal Uses

```python
# 1. Tree Deletion (delete nodes bottom-up)
def deleteTree(root):
    if not root:
        return
    
    deleteTree(root.left)
    deleteTree(root.right)
    # Delete current node after children
    del root

# 2. Calculate Tree Height/Depth
def maxDepth(root):
    if not root:
        return 0
    
    left_height = maxDepth(root.left)
    right_height = maxDepth(root.right)
    return max(left_height, right_height) + 1

# 3. Tree Size/Sum Calculations
def treeSum(root):
    if not root:
        return 0
    
    left_sum = treeSum(root.left)
    right_sum = treeSum(root.right)
    return root.val + left_sum + right_sum

# 4. Lowest Common Ancestor
def lowestCommonAncestor(root, p, q):
    if not root or root == p or root == q:
        return root
    
    left = lowestCommonAncestor(root.left, p, q)
    right = lowestCommonAncestor(root.right, p, q)
    
    if left and right:
        return root
    return left or right
```
