### Missing Number
#### Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.

```python
def missingNumber(nums):
    seen = set(nums)
    n = max(seen)
    
    i = 0
    while i < n:
        if i not in seen:
            return i
        i += 1
```

```python
nums = [3,0,1]
missingNumber(nums)
```

### Counting Elements
#### Given an integer array arr, count how many elements x there are, such that x + 1 is also in arr. If there are duplicates in arr, count them separately.

```python
def countElements(arr):
    count = 0
    seen = set(arr)
    
    for num in arr:
        if (num + 1) in seen:
            count += 1
    return count
        
```

### 1436. Destination City
#### You are given the array paths, where paths[i] = [cityAi, cityBi] means there exists a direct path going from cityAi to cityBi. Return the destination city, that is, the city without any path outgoing to another city.

#### It is guaranteed that the graph of paths forms a line without any loop, therefore, there will be exactly one destination city.

```python
def destCity(paths):
    seen = set()
    
    for path in paths:
        arrive, _ = path
        seen.add(arrive)
        
    for path in paths:
        _, depart = path
        if depart not in seen:
            return depart
    return None
```

```python
paths = [["London","New York"],["New York","Lima"],["Lima","Sao Paulo"]]
destCity(paths)
```

### 1496. Path Crossing
#### Given a string path, where path[i] = 'N', 'S', 'E' or 'W', each representing moving one unit north, south, east, or west, respectively. You start at the origin (0, 0) on a 2D plane and walk on the path specified by path.

#### Return true if the path crosses itself at any point, that is, if at any time you are on a location you have previously visited. Return false otherwise.

```python
def isPathCrossing(path):
    location = {0 : 0}
    direction = {
        'N': (0, 1),
        'E': (1, 0),
        'S': (0, -1),
        'W': (-1, 0)
    }
    
    for coord in path:
        x, y = direction[coord]
        if x in location:
            if location[x] == y:
                return True
        else:
            location[x] = y
    return False
        
```

```python
path = "NES"
isPathCrossing(path)
```

### Find Players With Zero or One Losses
#### You are given an integer array matches where matches[i] = [winneri, loseri] indicates that the player winneri defeated player loseri in a match.

#### Return a list answer of size 2 where:

#### answer[0] is a list of all players that have not lost any matches.
#### answer[1] is a list of all players that have lost exactly one match.
#### The values in the two lists should be returned in increasing order.

#### Note:

#### You should only consider the players that have played at least one match.
#### The testcases will be generated such that no two matches will have the same outcome.

```python
def findWinners(matches):
    seen = set()
    losses = {}
    
    for match in matches:
        winner, loser = match
        seen.add(winner)
        seen.add(loser)
        
        if loser in losses:
            losses[loser] += 1
        else:
            losses[loser] = 1
    
    losers = []
    for player in losses:
        if losses[player] == 1:
            losers.append(player)
        seen.remove(player)
    
    return [list(seen), losers]
        
```

```python
matches = [[1,3],[2,3],[3,6],[5,6],[5,7],[4,5],[4,8],[4,9],[10,4],[10,9]]
findWinners(matches)
```

### Largest Unique Number
#### Given an integer array nums, return the largest integer that only occurs once. If no integer occurs once, return -1.

```python
def largestUnique(nums):
    largest = float('-inf')
    freq = {}
    
    for val in nums:
        if val in freq:
            freq[val] += 1
        else:
            freq[val] = 1
    
    for num in freq:
        if freq[num] == 1:
            largest = max(largest, num)
    return largest
        
```

```python
nums = [5,7,3,9,4,9,8,3,1]
largestUnique(nums)
```

### Contiguous Array
#### Given a binary array nums, return the maximum length of a contiguous subarray with an equal number of 0 and 1.

```python
def findMaxLength(nums):
    max_len = float('-inf')
    location = {0: -1}
    
    count = 0
    for right in range(len(nums)):
        if nums[right] == 1:
            count += 1
        else:
            count -= 1
        # when same count appears twice, subarray has equal 0s and 1s.
        if count in location:
            curr_len = right - location[count]
            max_len = max(max_len, curr_len)
        else:
            location[count] = right
    return max_len
```

```python
nums = [0,1]
findMaxLength(nums)
```

### Ransom Note
#### Given two strings ransomNote and magazine, return true if ransomNote can be constructed by using the letters from magazine and false otherwise.

#### Each letter in magazine can only be used once in ransomNote.

```python
def canConstruct(ransomNote, magazine):
    hash = {}
    for char in magazine:
        if char in hash:
            hash[char] += 1
        else:
            hash[char] = 1
        
    for char in ransomNote:
        if char not in hash:
            return False
        else:
            hash[char] -= 1
    return True
```

```python
ransomNote = "a"
magazine = "b"
canConstruct(ransomNote, magazine)
```

### Longest Substring Without Repeating Characters
#### Given a string s, find the length of the longest substring without duplicate characters.

```python
def lengthOfLongestSubstring(s):
    max_len = float('-inf')
    seen = set(s)
    
    for char in s:
        if char in seen:
            seen.remove(char)
        else:
            seen.add(char)
        max_len = max(max_len, len(seen))
    return max_len
    
        
```

### 3005. Count Elements With Maximum Frequency
#### You are given an array nums consisting of positive integers.

#### Return the total frequencies of elements in nums such that those elements all have the maximum frequency.

#### The frequency of an element is the number of occurrences of that element in the array.

```python
def maxFrequencyElements(nums):
    max_freq = 0
    freq = {}
    
    for num in nums:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
            
    max_count = max(freq.values())
    for num in freq:
        if freq[num] == max_count:
            max_freq += max_count
    return max_freq
    
    
```

### 1207. Unique Number of Occurrences
#### Given an array of integers arr, return true if the number of occurrences of each value in the array is unique or false otherwise.

```python
def uniqueOccurrences(arr):
    freq = {}
    
    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    
    prev = 0
    for num in freq:
        if freq[num] == prev:
            return False
        prev = freq[num]
    return True
        
```

### 451. Sort Characters By Frequency
#### Given a string s, sort it in decreasing order based on the frequency of the characters. The frequency of a character is the number of times it appears in the string.

#### Return the sorted string. If there are multiple answers, return any of them.

```python
def frequencySort(s):
    sorted = ""
    freq = {}
    
    for num in s:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
            
    bucket = [[] for _ in range(len(s))]
    for char, count in freq.items():
        bucket[count].append(char)
    
    for freq in range(len(bucket)-1, -1, -1):
        if bucket[freq]:
            for char in bucket[freq]:
                sorted += char * freq
    return sorted
```

```python
s = "Aabb"
frequencySort(s)
```

### 2958. Length of Longest Subarray With at Most K Frequency
#### You are given an integer array nums and an integer k.

#### The frequency of an element x is the number of times it occurs in an array.

#### An array is called good if the frequency of each element in this array is less than or equal to k.

#### Return the length of the longest good subarray of nums.

#### A subarray is a contiguous non-empty sequence of elements within an array.

```python
def maxSubarrayLength(nums, k):
    max_len = float('-inf')
    freq = {}
    
    left = 0
    for right, num in enumerate(nums):
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
        
        while freq[num] > k:
            curr = nums[left]
            if curr in freq:
                freq[curr] -= 1
            left += 1
        max_len = max(max_len, right - left + 1)
    return max_len
```

```python
nums = [5,5,5,5,5,5,5]
k = 4
maxSubarrayLength(nums, k)
```

### 1512. Number of Good Pairs
#### Given an array of integers nums, return the number of good pairs.

#### A pair (i, j) is called good if nums[i] == nums[j] and i < j.

```python
def numIdenticalPairs(nums):
    num_identicals = 0
    count = {}
    
    for num in nums:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1
    
    for num in count:
        if count[num] == 2:
            num_identicals += 1
        elif count[num] > 2:
            num_identicals += count[num]
    return num_identicals
```

### 930. Binary Subarrays With Sum
#### Given a binary array nums and an integer goal, return the number of non-empty subarrays with a sum goal.

#### A subarray is a contiguous part of the array.

```python
def numSubarraysWithSum(nums, goal):
    prefix_count = {0: 1}
    count = 0
    
    curr = 0
    for num in nums:
        # subarray sum 
        curr += num
        
        # goal = curr - prev
        prev = curr - goal
        if prev in prefix_count:
            # how many subarrays exist
            count += prefix_count[prev]
            print(count)
            
        prefix_count[curr] = prefix_count.get(curr, 0) + 1
    return prefix_count
```

### 1695. Maximum Erasure Value
#### You are given an array of positive integers nums and want to erase a subarray containing unique elements. The score you get by erasing the subarray is equal to the sum of its elements.

#### Return the maximum score you can get by erasing exactly one subarray.

#### An array b is called to be a subarray of a if it forms a contiguous subsequence of a, that is, if it is equal to a[l],a[l+1],...,a[r] for some (l,r).

```python
def maximumUniqueSubarray(nums):
    max_sum = 0
    seen = set()
    
    left, curr = 0, 0
    for right in range(len(nums)):
        
        while nums[right] in seen:
            seen.remove(nums[left])
            curr -= nums[left]
            left += 1
        
        seen.add(nums[right])
        curr += nums[right]
        
        max_sum = max(max_sum, curr)
    return max_sum
            
```

```python
nums = [4,2,4,5,6]
maximumUniqueSubarray(nums)
```

### 567. Permutation in String
#### Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

#### In other words, return true if one of s1's permutations is the substring of s2.



```python
from collections import Counter

def checkInclusion(s1, s2):
    target = Counter(s1)
    window = Counter()
    
    n, m = len(s1), len(s2)
    
    if n > m:
        return False
    
    left = 0
    for right, char in enumerate(s2):
        window[char] += 1
        
        while (right - left + 1) > n:
            curr = s2[left]
            del window[curr]
            left += 1
        
        if window == target:
            return True
    return False
            
```

```python
s1 = "ab"
s2 = "eidbaoo"
checkInclusion(s1, s2)
```

### 205. Isomorphic Strings
#### Given two strings s and t, determine if they are isomorphic.

#### Two strings s and t are isomorphic if the characters in s can be replaced to get t.

#### All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

```python
from collections import Counter

def isIsomorphic(s, t):
    one = Counter(s)
    two = Counter(t)
    
    for target, count in zip(one.values(), two.values()):
        if count != target:
            return False
    return True
```

### 290. Word Pattern
#### Given a pattern and a string s, find if s follows the same pattern.

#### Here follow means a full match, such that there is a bijection between a letter in pattern and a non-empty word in s. Specifically:

#### Each letter in pattern maps to exactly one unique word in s.
#### Each unique word in s maps to exactly one letter in pattern.
#### No two letters map to the same word, and no two words map to the same letter.

```python
def wordPattern(pattern, s):
    seen = set(pattern)
    
    words = s.split(' ')
    target = set(words)
    
    if len(seen) != len(target):
        return False
    
    map = {}
    for char, word in zip(seen, words):
        map[char] = word
        
    return True
    
```

```python
pattern = "abba"
s = "dog cat cat fish"
wordPattern(pattern, s)
```

### 791. Custom Sort String
#### You are given two strings order and s. All the characters of order are unique and were sorted in some custom order previously.

#### Permute the characters of s so that they match the order that order was sorted. More specifically, if a character x occurs before a character y in order, then x should occur before y in the permuted string.

#### Return any permutation of s that satisfies this property.

```python
def customSortString(order, s):
    ans = ''
    seen = set(s)
    
    for char in order:
        if char in seen:
            ans += char
            seen.remove(char)
            
    left = ''.join(list(seen))
    ans += left
    
    return ans
```

```python
order = "bcafg"
s = "abcd"
customSortString(order, s)
```

### 1657. Determine if Two Strings Are Close
#### Two strings are considered close if you can attain one from the other using the following operations:

#### Operation 1: Swap any two existing characters.
#### For example, abcde -> aecdb
#### Operation 2: Transform every occurrence of one existing character into another existing character, and do the same with the other character.
#### For example, aacabb -> bbcbaa (all a's turn into b's, and all b's turn into a's)
#### You can use the operations on either string as many times as necessary.

#### Given two strings, word1 and word2, return true if word1 and word2 are close, and false otherwise.

```python
from collections import Counter

def closeStrings(word1, word2):
    one = Counter(word1)
    two = Counter(word2)
    
    for freq, count in zip(one.values(), two.values()):
        if freq != count:
            return False
    return True
```

```python
word1 = "cabbba"
word2 = "abbccc"
closeStrings(word1, word2)
```
