## Two Pointers

```python
def reverseWord(s):
    sent = s.split(' ')
    reverse = []
    
    for word in sent:
        left = 0
        right = len(word) - 1
        word = list(word)
        while left < right:
            word[left], word[right] = word[right], word[left]
            left += 1
            right -= 1
        reverse.append(''.join(word))
    
    return ' '.join(reverse)
```

```python
def reverseWord(s):
    chars = list(s)
    
    space = 0
    for i in range(len(chars) + 1):
        if i == len(chars) or chars[i] == ' ':
            left, right = space, i - 1
            while left < right:
                chars[left], chars[right] = chars[right], chars[left]
                left += 1
                right -= 1
            space = i + 1
    return ''.join(chars)
```

```python
sent = reverseWord(s)
```

```python
def reverseOnlyLetters(s):
    # stack of characters    
    letters = [char for char in s if char.isalpha()]
    ans = []
    
    for char in s:
        if char.isalpha():
            # reverse letters
            ans.append(letters.pop())
        else:
            ans.append(char)
    
    
    return "".join(ans)
```

```python
s = "a-bC-dEf-ghIj"
```

```python
sent = reverseOnlyLetters(s)
```

```python
sent
```

```python
def getCommon(nums1, nums2):
    ptr1 = 0
    ptr2 = 0
    
    while ptr1 < len(nums1) and ptr2 < len(nums2):
        if nums1[ptr1] < nums2[ptr2]:
            ptr1 += 1
        elif nums1[ptr1] > nums2[ptr2]:
            ptr2 += 1
        else:
            return nums1[ptr1]
    
    return - 1
```

```python
nums1 = [1,2,3]
nums2 = [2,4]
```

```python
ans = getCommon(nums1, nums2)
```

```python
ans
```

```python
def moveZeroes(nums):
    left = 0
    
    for right in range(len(nums)):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
    
    return nums
```

```python
nums = [0,1,0,3,12]
```

```python
print(moveZeroes(nums))
```

```python
def reversePrefix(word, char):
    result = list(word)
    left = 0
    
    for right in range(len(word)):
        if word[right] == char:
            while left < right:
                result[right], result[left] = result[left], result[right]
                left += 1
                right -= 1
            return ''.join(result)
    return word
```

```python
word = "abcdefd"
ch = "d"
```

```python
ans = reversePrefix(word, ch)
```

```python
ans
```

## Sliding Window

```python
def minSubArrayLen(target, nums):
    left = 0
    curr = 0
    
    min_len = float('inf')
    for right in range(len(nums)):  
        curr += nums[right]
        
        while curr >= target:
            min_len = min(min_len, right - left + 1)
            curr -= nums[left]
            left += 1
            
    return min_len if min_len != float('inf') else 0
```

```python
target = 7
nums = [2,3,1,2,4,3]
```

```python
ans = minSubArrayLen(target, nums)
```

```python
ans
```

```python
def maxVowels(s, k):
    vowels = set('aeiou')
    count = 0
    max_vowels = float('-inf')
    
    left = 0
    for right in range(len(s)):
        if s[right] in vowels:
            count += 1
        
        # shrink window size
        while right - left + 1 > k:
            if s[left] in vowels:
                count -= 1
            left += 1
        
        max_vowels = max(max_vowels, count)
    
    return max_vowels if max_vowels != float('-inf') else 0
            
```

```python
s = "abciiidef"
k = 3
```

```python
maxVowels(s, k)
```

```python
def equalSubstring(s, t, maxCost):
    max_len = float('-inf')
    curr_cost = 0
    
    left = 0
    for right in range(len(s)):
        curr_cost += abs(ord(s[right]) - ord(t[right]))
        
        while curr_cost > maxCost:
            curr_cost -= abs(ord(s[left]) - ord(t[left]))
            left += 1
        
        max_len = max(max_len, right - left + 1)
    
    return max_len if max_len != float('-inf') else 0
        
```

```python
s = "abcd"
t = "bcdf"
maxCost = 3
```

```python
equalSubstring(s, t, maxCost)
```

## Prefix Sum

```python
def fn(arr):
    prefix = [arr[0]]
    for i in range(1, len(arr)):
        prefix.append(prefix[-1] + arr[i])
    
    return prefix
```

```python
def largestAltitude(gain):
    highest = 0
    
    curr = 0
    for i in range(len(gain)):
        curr += gain[i]
        highest = max(highest, curr)
    
    return highest
```

```python
gain = [-4,-3,-2,-1,4,3,2]
```

```python
largestAltitude(gain)
```

```python
def pivotIndex(nums):
    prefix = [nums[0]]
    for i in range(1, len(nums)):
        prefix.append(prefix[-1] + nums[i])
        
    total = prefix[-1]
    for i in range(len(nums)):
        left = prefix[i-1] if i > 0 else 0
        right = total - prefix[i]
        
        if left == right:
            return i
    
    return -1
```

```python
def pivotIndex(nums):
    total = sum(nums)
    
    curr = 0
    for i in range(len(nums)):
        if curr == total - curr - nums[i]:
            return i
        curr += nums[i]
        
    return -1
```

```python
nums = [1,7,3,6,5,6]
```

```python
pivotIndex(nums)
```
